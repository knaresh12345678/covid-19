# STEP 1 → Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
warnings.filterwarnings("ignore")
```

# STEP 2 → Load Dataset

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['COVID clinical trials.csv']))
df.head()
```

```python
df.info()
df.isnull().sum()
df.describe(include="all")
```

# STEP 3 -> CLEAN DATA

```python
# Convert dates
date_cols = ['Start Date', 'Primary Completion Date', 'Completion Date']
for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors='coerce')

# Fill missing categorical values
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(f"Missing_{col}")

# Fill numeric missing
if 'Enrollment' in df.columns:
    df['Enrollment'] = df['Enrollment'].fillna(df['Enrollment'].median())
```

# STEP 4 -> RESEARCH QUESTION **1:** Trial Duration Analysis   (How long do COVID-19 clinical trials take to complete, and which countries complete trials the fastest?

CELL 1 — Create Duration Column

```python
df['Duration_days'] = (df['Completion Date'] - df['Start Date']).dt.days
duration_df = df[df['Duration_days'] > 0]

print("Q1 STEP 1: Duration column created successfully.")
print("Valid trials with duration:", len(duration_df))
```

CELL 2 — Country-wise Median Duration

```python
country_duration = duration_df.groupby('Country')['Duration_days'].median().sort_values()

print("Q1 STEP 2: Median trial duration calculated for each country.")
print("Number of countries in dataset:", len(country_duration))
print("\nSample of first 10 countries:\n")
print(country_duration.head(10))
```

CELL 3 — Fastest 10 Countries (Graph + Print Answer)

```python
fastest = country_duration.head(10)

plt.figure(figsize=(12,6))
fastest.plot(kind='barh', color='green')
plt.title("Top 10 Fastest Countries (Median Trial Duration)")
plt.xlabel("Median Duration (days)")
plt.ylabel("Country")
plt.show()

print("Q1 STEP 3: Fastest 10 Countries:")
print(fastest)
```

CELL 4 — Slowest 10 Countries (Graph + Print Answer)

```python
slowest = country_duration.tail(10)

plt.figure(figsize=(12,6))
slowest.plot(kind='barh', color='red')
plt.title("Top 10 Slowest Countries (Median Trial Duration)")
plt.xlabel("Median Duration (days)")
plt.ylabel("Country")
plt.show()

print("Q1 STEP 4: Slowest 10 Countries:")
print(slowest)
```

CELL 5 — Final Interpretation (Print Answer)

```python
fastest_country = fastest.index[0]
fastest_value = fastest.iloc[0]

slowest_country = slowest.index[-1]
slowest_value = slowest.iloc[-1]

print("Q1 FINAL INTERPRETATION:\n")
print(f"✔ Fastest Country: {fastest_country} (Median duration: {fastest_value} days)")
print(f"✔ Slowest Country: {slowest_country} (Median duration: {slowest_value} days)\n")


```

# STEP 5 -> Research Question 2– **How** did intervention types change over time?


CELL 1 — Extract Month from Start Date

```python
df['Month'] = df['Start Date'].dt.to_period('M')

print("Q2 STEP 1: Month column created successfully.")
print(df[['Start Date', 'Month']].head())
```

CELL 2 — Standardize Intervention Type

```python
def categorize_intervention(text):
    text = str(text).lower()
    if "vaccine" in text:
        return "Vaccine"
    elif "drug" in text:
        return "Drug"
    elif "diagnostic" in text or "test" in text:
        return "Diagnostic"
    elif "biological" in text:
        return "Biological"
    else:
        return "Other"

df['Intervention_Type'] = df['Interventions'].apply(categorize_intervention)

print("Q2 STEP 2: Intervention_Type column created.")
print(df['Intervention_Type'].value_counts())
```

CELL 3 — Group by Month and Intervention Type

```python
monthly_trends = df.groupby(['Month', 'Intervention_Type']).size().unstack().fillna(0)

print("Q2 STEP 3: Monthly trends table created.")
monthly_trends.head()
```

CELL 4 — Plot Intervention Trends Over Time

```python
plt.figure(figsize=(14,7))
for col in monthly_trends.columns:
    plt.plot(monthly_trends.index.to_timestamp(),
             monthly_trends[col], marker='o', label=col)

plt.title("Intervention Trends Over Time", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Number of Trials")
plt.legend(title="Intervention Type")
plt.grid(True)
plt.show()

print("Q2 STEP 4: Improved Line Chart displayed.")
```

CELL 5 — Interpretation (Printed Answer)

```python
print("Q2 FINAL INTERPRETATION:\n")

top_month = monthly_trends.sum(axis=1).idxmax()
print(f"- The month with the highest number of trials is: {top_month}")


```

# STEP 6.Research Question 3– Do wealthier countries conduct more advanced-phase clinical trials?


CELL 1 — Create Income Mapping

```python
income_map = {
    "United States": "High",
    "France": "High",
    "Germany": "High",
    "India": "Lower-Middle",
    "China": "Upper-Middle",
    "Brazil": "Upper-Middle",
    "Nigeria": "Low",
    "Pakistan": "Lower-Middle",
}

df['IncomeLevel'] = df['Country'].map(income_map).fillna("Unknown")

print("Q3 STEP 1: Income level assigned to countries.")
print(df[['Country', 'IncomeLevel']].head())
```

CELL 2 — Group Income vs Phase

```python
income_phase = df.groupby(['IncomeLevel', 'Phases']).size().unstack().fillna(0)

print("Q3 STEP 2: Income level vs trial phase table created.")
income_phase
```

CELL 3 — Plot Income Level vs Phase (Stacked Bar)

```python
plt.figure(figsize=(14,7))
income_phase.plot(kind='bar', stacked=True, figsize=(14,7))

plt.title("Income Level vs Trial Phase")
plt.xlabel("Income Level")
plt.ylabel("Number of Trials")
plt.legend(title="Trial Phase")
plt.show()

print("Q3 STEP 3: Stacked bar chart displayed.")
```

CELL 4 — Interpretation

```python
advanced_phases = income_phase[['Phase 3', 'Phase 4']].sum(axis=1).sort_values(ascending=False)
early_phases = income_phase[['Phase 1', 'Phase 2']].sum(axis=1).sort_values(ascending=False)

top_income_group = advanced_phases.index[0]
top_value = advanced_phases.iloc[0]

early_group = early_phases.index[0]
early_value = early_phases.iloc[0]

print("Q3 FINAL INTERPRETATION:\n")
print(f"- {top_income_group} income countries conducted the highest number of advanced-phase trials ({top_value}).")
print(f"- {early_group} income countries conducted the most early-phase trials ({early_value}).")

```

# STEP 12 -> Research Question 9: Do vaccine trials complete faster than drug

CELL 1 — Categorize Intervention Type for Drugs & Vaccines

```python
def intervention_category(text):
    text = str(text).lower()
    if "vaccine" in text:
        return "Vaccine"
    elif "drug" in text:
        return "Drug"
    else:
        return "Other"

df['Intervention_Class'] = df['Interventions'].apply(intervention_category)

valid_df = df[(df['Intervention_Class'].isin(['Drug','Vaccine'])) & (df['Duration_days'] > 0)]

print("Q9 STEP 1: Drug and Vaccine trials extracted:", len(valid_df))
```

CELL 2 — Plot Duration Comparison

```python
plt.figure(figsize=(12,6))
sns.histplot(valid_df[valid_df['Intervention_Class']=='Drug']['Duration_days'], kde=True, color='orange', label='Drug', alpha=0.5)
sns.histplot(valid_df[valid_df['Intervention_Class']=='Vaccine']['Duration_days'], kde=True, color='green', label='Vaccine', alpha=0.5)

plt.title("Drug vs Vaccine Trial Duration Distribution")
plt.xlabel("Duration (Days)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

print("Q9 STEP 2: Histogram comparison displayed.")
```

CELL 3 — Calculate Median Duration for Each Group

```python
median_duration = valid_df.groupby('Intervention_Class')['Duration_days'].median()

drug_duration = median_duration.get('Drug', None)
vaccine_duration = median_duration.get('Vaccine', None)

print("Q9 STEP 3: Median durations:")
print(median_duration)
print("Q9 FINAL INTERPRETATION:\n")
```
